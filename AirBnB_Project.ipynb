```python
# Calculate the cumulative sum of 'price' over time
abb_df['cumulative_price'] = abb_df['price'].cumsum()


# Plot the cumulative price over time
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
plt.plot(abb_df['last_review'], abb_df['cumulative_price'], color='cyan')
plt.xlabel('Date')
plt.ylabel('cumulative_price')
plt.title('Growth Over Time')
plt.grid(True)
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

```python
traffic_areas = abb_df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights', ascending=False)
traffic_areas
```

```python
room_type = traffic_areas['room_type']
stayed = traffic_areas['minimum_nights']

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(room_type, stayed, color ='violet',
        width = 0.4)

plt.xlabel("Room Type")
plt.ylabel("Minimum number of nights stayed")
plt.title("Traffic Areas")
plt.show()
```

```python
import urllib.request
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import io

background_image_url = 'https://raw.githubusercontent.com/lewagon/data-images/master/workshops/Neighbourhoods_New_York_City_Map.png'

# Open the URL and read the image data.
with urllib.request.urlopen(background_image_url) as url_obj:
    image_data = url_obj.read()

# Wrap the image data in a io.BytesIO object.
image_file = io.BytesIO(image_data)

# Convert the image data to a NumPy array.
background_image = np.array(Image.open(image_file))

# Plot the background image.
plt.imshow(background_image, zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])

# Plot the affordable housing data.
ax = plt.gca()
affordable_df.plot(
    ax=ax,
    zorder=1,
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8)
)

plt.show()
```
